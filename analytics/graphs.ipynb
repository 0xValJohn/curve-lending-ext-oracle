```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

data = pd.read_csv('dataset.csv')
```

```python
# Create a trace for the st_yeth_crvusd
trace1 = go.Scatter(
    x=data['Timestamp'],
    y=data['st_yeth_crvusd_no_chainlink'],
    mode='lines+markers',
    name='st_yeth_crvusd_no_chainlink'  # Add a name for the legend
)

trace2 = go.Scatter(
    x=data['Timestamp'],
    y=data['st_yeth_crvusd_with_chainlink'],
    mode='lines+markers',
    name='st_yeth_crvusd_with_chainlink'  # Add a name for the legend
)

# Create a trace for the eth_crvusd
trace3 = go.Scatter(
    x=data['Timestamp'],
    y=data['eth_crvusd'],
    mode='lines+markers',
    name='eth_crvusd'  # Add a name for the legend
)

# Create a layout for the graph
layout = go.Layout(
    title='st-yETH/crvUSD Oracle Price over time',
    xaxis=dict(title='Timestamp'),
    yaxis=dict(title='Value'),
    legend=dict(x=0.1, y=1.1)  # Add a legend
)

# Create the figure and display it
fig = go.Figure(data=[trace1, trace2, trace3], layout=layout)
fig.show()
```
